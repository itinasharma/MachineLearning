```python
import numpy as np
from sklearn.decomposition import PCA

# A 3x2 matrix: maps 2D inputs to 3D outputs, but rank is only 2
# So all outputs land on a 2D plane inside 3D space
W = np.array([[1., 0.],
              [0., 1.],
              [1., 1.]])   # third row = sum of first two, so rank 2

# Generate 1000 random 2D inputs and transform them
X = np.random.randn(1000, 2)
Y = X @ W.T   # shape (1000, 3)

# PCA on the outputs: how many dimensions hold the variance?
pca = PCA(n_components=3)
pca.fit(Y)
print("Explained variance by component:")
for i, v in enumerate(pca.explained_variance_ratio_):
    print(f"  Component {i+1}: {v:.4f}")
```
