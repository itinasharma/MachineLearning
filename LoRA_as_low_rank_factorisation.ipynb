```python
import numpy as np

def inspect_matrix(W, name="W"):
    u, s, vt = np.linalg.svd(W, full_matrices=False)
    rank = np.sum(s > 1e-8)
    m, n = W.shape
    print(f"{name}  shape={W.shape}")
    print(f"  rank={rank}  (max possible: {min(m,n)})")
    print(f"  top-5 singular values: {np.round(s[:5], 4)}")
    print()

m = n = 64   # toy stand-in for 4096x4096
r = 4        # rank of the LoRA update

np.random.seed(0)
A = np.random.randn(m, r)   # tall and thin
B = np.random.randn(r, n)   # short and wide
Delta = A @ B               # the low-rank update

inspect_matrix(Delta, "LoRA Delta (A @ B)")

full_params = m * n
lora_params = m * r + r * n
print(f"Full matrix params: {full_params:,}")
print(f"LoRA params:        {lora_params:,}")
print(f"Reduction:          {full_params // lora_params}x")
```
