```python
import numpy as np

def inspect_matrix(W, name="W"):
  # Singular Value Decomposition
    # Non-zero singular values tell us the rank
    u, s, vt = np.linalg.svd(W)
     # Small tolerance for floating point errors
    eps = 1e-8
    rank = np.sum(s > eps)
    m, n = W.shape
    # For an m x n matrix:
    # nullity = number of destroyed input directions
    nullity = n - rank
    print(f"{name}  shape={W.shape}")
    print(f"  singular values = {np.round(s, 4)}")
    print(f"  rank={rank}  max_possible={min(m,n)}  nullity={nullity}")
    print()


# A pure rotation.
# Points move, but no information is lost.
theta = np.deg2rad(45)
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
inspect_matrix(R, "Rotation R")

# Every output lies on a single line.
# Second row is just 2x the first row -> rank collapses to 1.
S = np.array([[1., 1.],
              [2., 2.]])
inspect_matrix(S, "Squish S")
```
