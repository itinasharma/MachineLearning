```python
import numpy as np

def inspect_matrix(W, name="W"):
    u, s, vt = np.linalg.svd(W, full_matrices=False)

    eps = 1e-8
    rank = np.sum(s > eps)
    nullity = W.shape[1] - rank

    print(f"{name} shape={W.shape} rank={rank} nullity={nullity}")

def nullspace(W, eps=1e-8):
    u, s, vt = np.linalg.svd(W, full_matrices=True)

    rank = np.sum(s > eps)

    # Null-space basis vectors
    return vt[rank:].T

np.random.seed(0)

W = np.random.randn(64, 512)

inspect_matrix(W, "W (64x512)")

N = nullspace(W)

print(f"Null space basis shape: {N.shape}")

x_null = N[:, 0]

residual = np.linalg.norm(W @ x_null)

print(f"norm(W @ x_null) = {residual:.2e}")
```
